```python
#Ejemplo: Si llueve, entonces llevo paraguas

# Definición de la función de implicación (p -> q)
# En lógica de programación: p implica q es equivalente a (not p or q)

def implicacion(p, q):
    return (not p) or q

# Escenarios posibles (Tabla de verdad)
# Cada tupla representa (llueve, llevo_paraguas)
escenarios = [
    (True, True),   # Caso 1: Llueve y llevo paraguas
    (True, False),  # Caso 2: Llueve y NO llevo paraguas (Único caso Falso)
    (False, True),  # Caso 3: No llueve y llevo paraguas
    (False, False)  # Caso 4: No llueve y no llevo paraguas
]

print(f"{'LLUEVE (p)':<12} | {'PARAGUAS (q)':<15} | {'RESULTADO (p -> q)'}")
print("-" * 50)

for llueve, paraguas in escenarios:
    resultado = implicacion(llueve, paraguas)

    # Formateo de salida para que sea legible
    p_str = str(llueve)
    q_str = str(paraguas)
    res_str = "VERDADERO" if resultado else "FALSO"

    print(f"{p_str:<12} | {q_str:<15} | {res_str}")

```

```python
#Ejemplo: No es cierto que me gusta la pizza

# Definición de la proposición simple
# p = "Me gusta la pizza"

def tabla_verdad_negacion():
    # Escenarios posibles para p (Verdadero o Falso)
    valores_p = [True, False]

    print(f"{'PIZZA (p)':<12} | {'NO PIZZA (¬p)':<15}")
    print("-" * 30)

    for p in valores_p:
        # El operador 'not' en Python es la negación lógica
        no_p = not p

        # Formateo para visualización
        p_str = "SÍ" if p else "NO"
        no_p_str = "VERDADERO" if no_p else "FALSO"

        print(f"{p_str:<12} | {no_p_str:<15}")

if __name__ == "__main__":
    print("PROPOSICIÓN: 'No es cierto que me gusta la pizza'\n")
    tabla_verdad_negacion()
```

```python
# Definición de la disyunción (p o q)
def disyuncion(p, q):
    return p or q

# Escenarios posibles (Tabla de verdad)
# p = Voy al cine
# q = Me quedo en casa
escenarios = [
    (True, True),   # Caso 1: Hago ambas cosas (ej. voy al cine y luego vuelvo a casa)
    (True, False),  # Caso 2: Solo voy al cine
    (False, True),  # Caso 3: Solo me quedo en casa
    (False, False)  # Caso 4: No hago ninguna de las dos (Único caso FALSO)
]

print(f"{'CINE (p)':<12} | {'CASA (q)':<12} | {'¿ES VERDAD? (p v q)'}")
print("-" * 45)

for p, q in escenarios:
    resultado = disyuncion(p, q)

    # Formateo para visualización
    p_str = "SÍ" if p else "NO"
    q_str = "SÍ" if q else "NO"
    res_str = "VERDADERO" if resultado else "FALSO"

    print(f"{p_str:<12} | {q_str:<12} | {res_str}")
```
